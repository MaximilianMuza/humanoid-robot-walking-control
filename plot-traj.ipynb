```python
import pandas as pd

motion_traj = pd.read_csv("homework-4/MJ_foot_graph.csv")
motion_traj.columns = ['walking_tick', 'left_foot_x','left_foot_y','left_foot_z', 'right_foot_x','right_foot_y','right_foot_z']
motion_traj['time'] = motion_traj.index*0.005
```

#### Plot all in one plot

```python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

data = motion_traj
time = data['time']
left_foot_color = 'blue'
right_foot_color = 'red'

fig = make_subplots(rows=3, 
                    cols=1,  # Remove the 0 before cols=1
                    vertical_spacing=0.2, 
                    subplot_titles=('X-Direction', 'Y-Direction', 'Z-Direction'),
                    shared_xaxes=True,
                    shared_yaxes=True
)
fig.add_trace(go.Scatter(x=time, y=data['x_pelv'], mode='lines', line=dict(color=left_foot_color, dash='dash'), showlegend=False), row=1, col=1)

fig.add_trace(go.Scatter(x=time, y=data['y_pelv'], mode='lines', line=dict(color=left_foot_color, dash='dash'), showlegend=False), row=2, col=1)

fig.add_trace(go.Scatter(x=time, y=data['z_pelv'], mode='lines', line=dict(color=left_foot_color, dash='dash'), showlegend=False), row=3, col=1)

fig.update_xaxes(title_text='Time [s]', row=3, col=1)
fig.update_yaxes(title_text='Displacement [m]', row=2, col=1)
fig.update_xaxes(dtick=1)
fig.update_layout(title='Pelvis Trajectories')
fig.show()
```

```python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

data = motion_traj
time = data['time']
left_foot_color = 'blue'
right_foot_color = 'red'

fig = make_subplots(rows=3, 
                    cols=1,  # Remove the 0 before cols=1
                    vertical_spacing=0.2,
                    subplot_titles=('X-Direction', 'Y-Direction', 'Z-Direction'),
                    shared_xaxes=True,
                    shared_yaxes=True
)
fig.add_trace(go.Scatter(x=time, y=data['left_foot_x'], mode='lines', line=dict(color=left_foot_color, dash='dash'), name='Left foot'), row=1, col=1)
fig.add_trace(go.Scatter(x=time, y=data['right_foot_x'], mode='lines', line=dict(color=right_foot_color), name='Right foot'), row=1, col=1)

fig.add_trace(go.Scatter(x=time, y=data['left_foot_y'], mode='lines', line=dict(color=left_foot_color, dash='dash'), showlegend=False, name='Left foot'), row=2, col=1)
fig.add_trace(go.Scatter(x=time, y=data['right_foot_y'], mode='lines', line=dict(color=right_foot_color), showlegend=False, name='Right foot'), row=2, col=1)

fig.add_trace(go.Scatter(x=time, y=data['left_foot_z'], mode='lines', line=dict(color=left_foot_color, dash='dash'), showlegend=False, name='Left foot'), row=3, col=1)
fig.add_trace(go.Scatter(x=time, y=data['right_foot_z'], mode='lines', line=dict(color=right_foot_color), showlegend=False, name='Right foot'), row=3, col=1)

fig.update_xaxes(title_text='Time [s]', row=3, col=1)
fig.update_yaxes(title_text='Displacement [m]', row=2, col=1)
fig.update_xaxes(dtick=1)
fig.update_layout(title='Foot Trajectories')
fig.show()
```
