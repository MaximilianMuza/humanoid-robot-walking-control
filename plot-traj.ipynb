```python
import pandas as pd

motion_traj = pd.read_csv("/home/maximilian/Projects/hum-rob/homework-2/motion_data.csv")
motion_traj.columns = ['walking_tick', 'curr_step', 'x', 'y']
motion_traj['time'] = motion_traj.index*0.005
```

#### Plot all in one plot

```python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

data = motion_traj
time = data['time']

fig = make_subplots(rows=2, cols=1, subplot_titles=('Left Foot', 'Right Foot'))
fig.add_trace(go.Scatter(x=time, y=data['x'], mode='lines', name='x-zmp'), row=1, col=1)
fig.add_trace(go.Scatter(x=time, y=data['y'], mode='lines', name='y-zmp'), row=2, col=1)
# fig.add_trace(go.Scatter(x=time, y=data['curr_step'], mode='lines', name='lfoot_x'), row=3, col=1)
# fig.add_trace(go.Scatter(x=time, y=data['lfoot_z'], mode='lines', name='lfoot_z'), row=2, col=1)

fig.update_xaxes(title_text='Time [s]', row=1, col=1)
fig.update_xaxes(title_text='Time [s]', row=2, col=1)
fig.update_yaxes(title_text='Position [m]', row=1, col=1)
fig.update_yaxes(title_text='Position [m]', row=2, col=1)
fig.update_layout(title='Foot Trajectories')
fig.show()
```

```python
import plotly.io as pio

pio.write_image(fig, 'total_trajectories.svg')
```
