```python
import pandas as pd

motion_traj = pd.read_csv("homework-3/motion_data.csv")
motion_traj.columns = ['walking_tick', 'curr_step', 'x', 'y']
motion_traj['time'] = motion_traj.index*0.005
```

#### Plot all in one plot

```python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

data = motion_traj
time = data['time']

fig = make_subplots(rows=2, cols=1, vertical_spacing=0.25)#, subplot_titles=('X', 'Y'))
fig.add_trace(go.Scatter(x=time, y=data['x'], mode='lines', name='x-zmp'), row=1, col=1)
fig.add_trace(go.Scatter(x=time, y=data['y'], mode='lines', name='y-zmp'), row=2, col=1)
# fig.add_trace(go.Scatter(x=time, y=data['curr_step'], mode='lines', name='lfoot_x'), row=3, col=1)
# fig.add_trace(go.Scatter(x=time, y=data['lfoot_z'], mode='lines', name='lfoot_z'), row=2, col=1)

fig.update_xaxes(title_text='Time [s]', row=1, col=1)
fig.update_xaxes(title_text='Time [s]', row=2, col=1)
fig.update_yaxes(title_text='Displacement [m]', row=1, col=1)
fig.update_yaxes(title_text='Displacement [m]', row=2, col=1)
fig.update_layout(title='ZMP Trajectory')
fig.show()
```

```python
import numpy as np
end = 2
t = np.linspace(0,end,100*end+1)

B = 1.0 / 2
t_d = 2
g = 9.81
omega_n = np.sqrt(g / )
T = 1
# K_x = B * t_d * omega_n / (t_d*omega_n + np.tanh(omega_n*(T/2 - t_d)))
K_x = 0.03
C_x1 = K_x - B
C_x2 = K_x / (t_d * omega_n)

x_com = np.zeros_like(t)
for i, t_i in enumerate(t):
    x_com[i] = C_x1 * np.cosh(omega_n * (t_i - t_d)) + C_x2 * np.sinh(omega_n * (t_i - t_d)) + B
```

```python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go


fig = make_subplots(rows=2, cols=1, vertical_spacing=0.25)#, subplot_titles=('X', 'Y'))
fig.add_trace(go.Scatter(x=t, y=x_com, mode='lines', name='x-zmp'), row=1, col=1)

fig.update_xaxes(title_text='Time [s]', row=1, col=1)
fig.update_xaxes(title_text='Time [s]', row=2, col=1)
fig.update_yaxes(title_text='Displacement [m]', row=1, col=1)
fig.update_yaxes(title_text='Displacement [m]', row=2, col=1)
fig.update_layout(title='ZMP Trajectory')
fig.show()
```
